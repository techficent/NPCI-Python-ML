### step 3: select features & target

```python
feature = apple_df[    ["Open"]    ]
target = apple_df[    ["Close"]        ]
```

### step 4: splitting the data

```python
x_train, x_test, y_train, y_test=train_test_split(    feature, target, test_size=0.2, random_state=42     )
```

### step 5: training the model

```python
model = LinearRegression()

model.fit(  x_train, y_train      )
```

### step 6: Evaluating the model

```python
x_test
```

```python
# Make a data frame of predicted close values
predicted_values_df = pd.DataFrame( model.predict(x_test), columns=["Predicted_Close"] )

#actual close values can be a data frame but their index is reset to avoid issues due to random sequence
actual_close_values_df = y_test.reset_index(drop=True)

#actual open values can be a data frame but their index is reset to avoid issues due to random sequence
actual_open_values_df = x_test.reset_index(drop=True)

#columns list to concatenate
cols = [actual_open_values_df, actual_close_values_df, predicted_values_df]

#use pd.concat to place columns SIDE BY SIDE , column wise [axis = 1]
result_df = pd.concat ( cols , axis=1     )

result_df
```

```python
result_df[["Close","Predicted_Close"]].corr()
```

```python
pe.line(     y=["Close","Predicted_Close"], data_frame=result_df    )
```

```python
print(      mean_squared_error(y_test, predicted_values_df, squared=False)    )
```
